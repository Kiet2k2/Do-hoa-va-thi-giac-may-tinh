```python
import cv2
import matplotlib.pyplot as plt

```

```python
anh = cv2.imread('pikachu.jpg')
anh2 = cv2.cvtColor(anh,cv2.COLOR_BGR2RGB)
plt.subplot(121)
plt.imshow(anh)
plt.subplot(122)
plt.imshow(anh2)
```

```python
anh.shape
anh_resize = cv2.resize(anh, (300,190), interpolation = cv2.INTER_AREA)
plt.subplot(121)
plt.imshow(anh)
plt.subplot(122)
plt.imshow(anh_resize)
```

```python
plt.imshow(anh2[71:88,50:110])
```

```python
#lm m nh
#Cch1
anh_filter = cv2.blur(anh2,(6,6))
plt.subplot(121)
plt.imshow(anh2)
plt.subplot(122)
plt.imshow(anh_filter)
```

```python
#Cch 2
anh_filter = cv2.boxFilter(anh_resize,-1,(7,7))
plt.subplot(121)
plt.imshow(anh2)
plt.subplot(122)
plt.imshow(anh_filter)
```
