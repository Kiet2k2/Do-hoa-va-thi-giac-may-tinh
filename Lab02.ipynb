```python

anh = cv2.imread('anna.jpg')
anh = anh[:,:,::-1]
#anh = cv2.cvtColor(anh,cv2.COLOR_RGB2GRAY)
anh_blur1 = cv2.blur(anh,(15,15))#ma trn cp 15  qut nh
anh_blur2 = cv2.GaussianBlur(anh,(15,15),0)
anh_blur3 = cv2.medianBlur(anh,9)#ch lc c s l
plt.subplot(141)
plt.imshow(anh)
plt.subplot(142)
plt.imshow(anh_blur1)
plt.subplot(143)
plt.imshow(anh_blur2)
plt.subplot(144)
plt.imshow(anh_blur3)
```
