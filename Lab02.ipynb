```python
#Ghép ảnh cùng ma trận cách 1
anhngang = cv2.hconcat([anh0[38:140,95:240],anh0[0:102,10:155]])
anhdoc = cv2.vconcat([anh0[38:140,95:240],anh0[0:102,10:155]])
plt.subplot(121)
plt.imshow(anhngang)
plt.subplot(122)
plt.imshow(anhdoc)
```

```python
anh1.shape
#chỉnh sửa kích thước của ảnh
anh3 = cv2.resize(anh2,(145,102))#145 là x ,102 là y
```

```python
#ghép ảnh sau khi chỉnh sửa anh2=anh3 cùng kích thước với ảnh 1
anhngang = cv2.hconcat([anh1,anh3])
anhdoc = cv2.vconcat([anh1,anh3])
plt.subplot(121)
plt.imshow(anhngang)
plt.subplot(122)
plt.imshow(anhdoc)
```

***Phần Filter bộ lọc ảnh***

```python

anh = cv2.imread('anna.jpg')
anh = anh[:,:,::-1]
#anh = cv2.cvtColor(anh,cv2.COLOR_RGB2GRAY)
anh_blur1 = cv2.blur(anh,(15,15))#ma trận cấp 15 để quét ảnh
anh_blur2 = cv2.GaussianBlur(anh,(15,15),0)
anh_blur3 = cv2.medianBlur(anh,9)#chỉ lọc được số lẻ
plt.subplot(141)
plt.imshow(anh)
plt.subplot(142)
plt.imshow(anh_blur1)
plt.subplot(143)
plt.imshow(anh_blur2)
plt.subplot(144)
plt.imshow(anh_blur3)
```
